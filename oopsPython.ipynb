```python
#create a constructor with methods

class Student:
    def __init__(self,name,age,grade): #constructor creation
        self.name=name
        self.age=age
        self.grade=grade

    def display_info(self): #method to display
        print(f"name:{self.name},\nage:{self.age},\ngrade:{self.grade}")

#now we create object of class
s=Student("Garima",20,"5th semester")
s.display_info()
```

```python
#inheritance example
class Animal:
    def speak(self):
        print("this is the parent class")

#to inherit from parent class we do
class Dog(Animal):
    def speak(self):  #overriding method
        print("woof! woof!")

class Cat(Animal):
    def speak(self):
        print("meow!")

#create objects
dog=Dog()
cat=Cat()
dog.speak()
cat.speak()
```

```python
#polymorphism example
#creating base class
class Shape:
    def area(self): #abstract method
        pass #means ignore, dont implement this method here
        
#creating child class
class Circle(Shape):
    def __init__(self,radius):
        self.radius=radius
        

    def area(self): #overriding area method
        return 3.14 * self.radius ** 2
        
    
#square class inherits shape
class Square(Shape):
    def __init__(self,length):
        self.length=length
        

    def area(self):
        return self.length ** 2
    
#demonstrating polymorphism
shapes=[Circle(5),Square(3)]
for shape in shapes:
    #print(f"Area : {shape.area()}")
    #here if i want to give the area name then i can do
    if isinstance(shape,Circle):
        print(f"The area of circle is {shape.area()}")
    elif isinstance (shape,Square):
        print(f"Area of a square: {shape.area()}")
```

```python
#decorator example
#A decoreator is just a function that takes another 
#function as input, adds extra behaviour and returns a new function
#defining a decorator function
def uppercase_decorator(func):
    def wrapper(): #inner method, which wraps or adds extra behaviour
        result=func() #calling the original function
        return result.upper() #modify result to uppercase
    return wrapper

#applying a decorator function
@uppercase_decorator
#it is a decorator syntax, 
#this means, greet=uppercase_decorator(greet)
def greet():
    return "hello world"

print(greet())
```

```python
#Encapsulation example
#it means restricting direct access to the variables

class BankAccount:
    def __init__(self,balance):
        self.__balance=balance
        #here double underscore to left side means private variable

    def deposit(self,amount):
        self.__balance +=amount 

    def withdraw(self,amount):
        if amount <= self.__balance:
            self.__balance -= amount
        else:
            print("Insufficient amount")

    def get_balance(self):
        return self.__balance #controlled access to private variable
    
    #example usage
account=BankAccount(200)
print("Current initial Balance:",account.get_balance())
#to show the initial balance

a=int(input("Enter an amount for deposit:"))
account.deposit(a)
print(f"Deposit amount: {a}")
print("Current Balance after deposit:",account.get_balance())
#to show the balance after deposit


b=int(input("Enter an amount for withdraw:"))
account.withdraw(b)
print(f"withdraw amount: {b}")
print("Good Balance after withdraw:",account.get_balance())
#to show the balance after withdraw
```
