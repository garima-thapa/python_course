```python
#sort a list without using built-in sort()

a=list(map(int,input("Enter a list of numbers separated by space").split()))
print(f"Numbers before sorting{a}")
for i in range(len(a)):
    for j in range(0,len(a)-i-1):
        if a[j] > a[j+1]:
            a[j],a[j+1]=a[j+1],a[j]
            

print("sorted list is:",a)
```

```python
#find the second largest and second smallest in the list

num=[22,34,55,11,66,54]

#finding second largest
largest=max(num)
num.remove(largest)
second_largest=max(num)
print("NUmbers in list are:",num)
print("Second largest number is:",second_largest)

#finding the second smallest number
smallest=min(num)
num.remove(smallest)
second_smallest=min(num)
print(f"Second smallest number is: {second_smallest}")
```

```python
#count frequencies of elements in numbers in a list
numbers=[1,2,3,3,3,2,2,1,5,6]
frequency={}

for num in numbers:
    frequency[num]=frequency.get(num,0)+1

print(f"Elements frequencies:{frequency}\n")
```

```python
#merge two lists and remove duplicates
list1=[1,2,3,4,5]
list2=[2,3,4,5,7]
merged_list=list1+list2
unique_list=list(set(merged_list))

print(f"Numbers without duplicates are {unique_list}")
```

```python
#implement stack and queue operations in list

#stack [LIFO]
stack=[]
stack.append(10)
stack.append(70)
stack.append(90)
print("stack afer pushes :",stack)
stack.pop()
print("stack after pop :",stack)

#Queue [FIFO]
queue=[2,3]
queue.append(77)
queue.append(8)
print("Queue after enqueues: ",queue)
queue.pop(0)
print("Queue after dequeue: ",queue)
```
